```python
import hashlib

def calculate_hash(file_path):
    with open(file_path, 'rb') as f:
        data = f.read()
        return hashlib.sha256(data).hexdigest()
```

```python
file = "Swap.txt"

hash_value = calculate_hash(file)
print("Original Hash:", hash_value)
```

```python
old_hash = calculate_hash("Swap.txt")

input("Modify the file and press Enter...")

new_hash = calculate_hash("Swap.txt")

if old_hash == new_hash:
    print("File is safe (No changes)")
else:
    print("Warning: File has been modified!")
```
