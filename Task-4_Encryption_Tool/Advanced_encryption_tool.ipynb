```python
# ============================================
# Advanced Encryption Tool (ULTIMATE )
# Multi-Key + Password + Hidden Folder
# ============================================

from cryptography.fernet import Fernet
import os
import hashlib
import platform

KEYS_DIR = ".secure_keys"

# -----------------------------------
# INIT (CREATE + HIDE FOLDER)
# -----------------------------------
def init():
    if not os.path.exists(KEYS_DIR):
        os.mkdir(KEYS_DIR)

        if platform.system() == "Windows":
            os.system(f'attrib +h {KEYS_DIR}')

# -----------------------------------
# HASH PASSWORD
# -----------------------------------
def hash_password(pwd):
    return hashlib.sha256(pwd.encode()).hexdigest()

# -----------------------------------
# CREATE KEY
# -----------------------------------
def generate_key():
    name = input("Enter key name: ")
    pwd = input("Set password: ")

    key = Fernet.generate_key()

    with open(f"{KEYS_DIR}/{name}.key", "wb") as f:
        f.write(key)

    with open(f"{KEYS_DIR}/{name}.pass", "w") as f:
        f.write(hash_password(pwd))

    key_hash = hashlib.sha256(key).hexdigest()
    with open(f"{KEYS_DIR}/{name}.hash", "w") as f:
        f.write(key_hash)

    print(f"[+] Key '{name}' created successfully")

# -----------------------------------
# VERIFY PASSWORD
# -----------------------------------
def verify_key(name):
    pwd = input("Enter key password: ")

    try:
        stored = open(f"{KEYS_DIR}/{name}.pass").read()
        return hash_password(pwd) == stored
    except:
        print("[-] Key not found")
        return False

# -----------------------------------
# CHECK KEY INTEGRITY
# -----------------------------------
def check_key_integrity(name, key):
    current = hashlib.sha256(key).hexdigest()
    stored = open(f"{KEYS_DIR}/{name}.hash").read()

    if current != stored:
        print("[!] WARNING: Key modified!")
        return False
    return True

# -----------------------------------
# LOAD KEY
# -----------------------------------
def load_key(name):
    return open(f"{KEYS_DIR}/{name}.key", "rb").read()

# -----------------------------------
# LIST KEYS
# -----------------------------------
def list_keys():
    keys = [f.replace(".key", "") for f in os.listdir(KEYS_DIR) if f.endswith(".key")]

    if not keys:
        print("[-] No keys found")
        return []

    print("\nAvailable Keys:")
    for i, k in enumerate(keys):
        print(f"{i+1}. {k}")

    return keys

# -----------------------------------
# LIST FILES
# -----------------------------------
def list_files():
    files = [f for f in os.listdir() if os.path.isfile(f)]

    print("\nAvailable Files:")
    for i, f in enumerate(files):
        print(f"{i+1}. {f}")

    return files

# -----------------------------------
# ENCRYPT FILE
# -----------------------------------
def encrypt_file():
    keys = list_keys()
    if not keys:
        return

    try:
        kchoice = int(input("Select key: ")) - 1
        key_name = keys[kchoice]
    except:
        print("[-] Invalid key selection")
        return

    if not verify_key(key_name):
        print("[-] Wrong password")
        return

    key = load_key(key_name)

    if not check_key_integrity(key_name, key):
        return

    fernet = Fernet(key)

    files = list_files()

    try:
        fchoice = int(input("Select file: ")) - 1
        filename = files[fchoice]
    except:
        print("[-] Invalid file selection")
        return

    with open(filename, "rb") as file:
        data = file.read()

    encrypted = fernet.encrypt(data)

    with open(filename, "wb") as file:
        file.write(encrypted)

    print(f"[+] File encrypted using key '{key_name}'")

# -----------------------------------
# DECRYPT FILE
# -----------------------------------
def decrypt_file():
    keys = list_keys()
    if not keys:
        return

    try:
        kchoice = int(input("Select key: ")) - 1
        key_name = keys[kchoice]
    except:
        print("[-] Invalid key selection")
        return

    if not verify_key(key_name):
        print("[-] Wrong password")
        return

    key = load_key(key_name)

    if not check_key_integrity(key_name, key):
        return

    fernet = Fernet(key)

    files = list_files()

    try:
        fchoice = int(input("Select file: ")) - 1
        filename = files[fchoice]
    except:
        print("[-] Invalid file selection")
        return

    try:
        with open(filename, "rb") as file:
            data = file.read()

        decrypted = fernet.decrypt(data)

        with open(filename, "wb") as file:
            file.write(decrypted)

        print(f"[+] File decrypted using key '{key_name}'")

    except:
        print("[-] Decryption failed")

# -----------------------------------
# MAIN MENU
# -----------------------------------
def main():
    init()

    while True:
        print("\n===== Advanced Encryption Tool =====")

        print("1. Create New Key")
        print("2. Encrypt File")
        print("3. Decrypt File")
        print("4. Exit")

        choice = input("Select option: ")

        if choice == "1":
            generate_key()
        elif choice == "2":
            encrypt_file()
        elif choice == "3":
            decrypt_file()
        elif choice == "4":
            break
        else:
            print("Invalid option")

# -----------------------------------
# RUN
# -----------------------------------
if __name__ == "__main__":
    main()
```
